# Utvecklingsmetodik

## Vattenfallsmetoden

ISO/IEC 12207 (finns på wikipedia). Gammal standard, kallas vattenfall. Används än i dag. Tar lång tid. Bra kvalitet på mjukvaran! Bra kvalitetssäkring. Enormt lång tid däremot, det är det stora problemet. T ex på Eriksson så tog det 18 månader att leverera, oavsett hur liten feature.  

2010 började stora företag ta sig bort från vattenfallsmetoden. Sen kom extremprogrammering, inga buggar fick gå olösta.  

## Kanban

Kanban kommer från Toyota i Japan. Började inom bilindustri men finns inom många områden i dag. Operations är drift, underhåll. Så när det driftas så är det ofta Kanban som används.  

I Kanban skriver vi massa lappar. De brukar vara tasks, user stories eller liknande. Lappen flyttas mellan olika rubriker (backlog, in progress, peer review, in test, done, blocked). Brukar vara möten varje dag när lappar flyttas, mötena kallas standup.  

Tanken med Kanban är att minska saker som bara fortsätter och fortsätter annars. Så alla lappar ska göras men varje person får bara ha ett maxantal lappar in progress (två eller en) på samma gång.  

T ex städa data inför AI/ML. Ta lappen från backlog, när du är klar sätter du den på peer review. När någon kollat på den så flyttas den vidare osv. Detta är alltså tänkt som en pipeline, går från ena till den andra och då ska det verkligen vara klart.  

Enkelt arbetssätt som visat sig vara väldigt effektivt. Enkelt att spåra vad som görs, vem som gör vad osv. Minskar också att många saker görs samtidigt.  

Det finns olika beskrivningar om alla metodiker.

## Scrum

Scrum är nog den största inom utveckling. Inte alls lika strukturerat som Kanban utan mer fritt. Ordet kommer från rugby (bildsök). Någon form av lagandavariant.  

Alla har roller. Scrummaster ser till att hålla möten, resten är utvecklare men i samband finns produktägare.  

Ska hållas dagliga möten. Finns någon produktbacklog. Nu är det verkligen user stories. En användardfokuserad berättelse. Backlogen är en prioriterad lista över saker som ska göras. Produktägaren prioriterar listan. Även produktägaren som skriver tasks och user stories men så fungerar det inte i praktiken, det sker i samarbete med teamet. Backlogen ska vara en beskrivning av allt arbete för att produkten ska kunna levereras.  

I scrum finns det alltid en backlog, så länge produkten existerar. Arbetet organiseras i sprinter, det vanligaste kanske är två veckor ungefär. En sprint är antal dagar då det planeras i detalj, tar sina user stories och bryter ner dem i små delar, sen spåras tiden (abstrakt tidsenhet, story points). Från början kanske teamet bara tittar på storleken, om de är large, medium eller small.  

Varje dag hålls en standup/daily. Mötet ska vara kort, 15 minuter. Ett statusmöte. Tre frågor:  

- vad gjorde jag i går?  
- vad vill jag göra i dag?  
- har jag några hinder?  

Efter mötet kan de tekniska detaljerna tas. Mötet är bara en avstämning för att få alla på samma blad över vad som ska göras för dagen.  

Det måste finnas en definition av vad det innebär att vara klar med en uppgift. Annars sitter du bara och jobbar och jobbar och hittar på uppgifter som kanske inte är rätt. Teamet vet inte när de är klara. Limit work in progress men det finns ingen formell gräns som i Kanban.  

I slutet av sprinten has en retrospective. Det pratas inte så mycket om vad som gick dåligt. Om problem finns kvar så är det en viktigt men hur saker gick till, vem gjorde vad osv det bryr inte scrum sig om. Vi pratar alltid om hur mycket tid som är kvar, inte hur mycket tid som gått åt.  

När du gjort story points några gånger så byter du från large, medium, small till kanske mer värden osv. Efter några sprintar i samma team så blir vi jättebra på att veta hur mycket vi hinner med. Det är i sig en sorts regressionsmodell. Tillslut kan vi kanske veta vad vi hinner på en halvdag. Sen räcker det med att en person byts ut så förändras hela dynamiken och alla värden och poäng. Värre kan vara att en manager tar beslut när tidsuppskattningen inte är mogen. Hårda deadlines undviks i scrum. Resultatet brukar inte bli och det blir massa crunch.  

Det stora problemet är att det kräver ganska mycket tillit. I andra metodiker så samlas alla och tar beslut i stora möten. I scrum sker saker lite hur som helst. Besluten ska tas i teamen, är tanken. Demand driven, inte supply driven. Allt arbete ska drivas av att en kund vill ha en feature, därav fokus på efterfrågan på marknaden. Mellanchefer och managers är inte riktigt nödvändiga då. Kan bli mycket motstånd över att införa scrum. Då finns inte tillit. Teamen är nästan självorganiserade. De synkar inte så mycket med varandra pga att det är dyrt. Produktägaren ska samordna. Väldigt dynamiskt. Allt i scrum är drivet av konversationer.  

## Safe development (scaled agile framework)

Inte något allmängiltigt. En produkt som ett företag säljer. Vattenfallprocess i scrumformat är vad det är. Tanken är att det ska vara skalbart. Stora företag tycker inte att scrum är skalbart (återigen tillitsfråga). Här hålls formella möten igen och utvärderar vad som hänt.
